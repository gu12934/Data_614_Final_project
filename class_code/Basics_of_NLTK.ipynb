## Basics of NLTK

The NLTK package allows for basic operations on text data. These are similar to what we did with the TM package in R, and include pre-processing of text. This file is based on these resources: https://okan.cloud/posts/2021-04-08-text-vectorization-using-python-term-document-matrix/,
https://realpython.com/nltk-nlp-python/, and
https://towardsdatascience.com/intro-to-nltk-for-nlp-with-python-87da6670dde

In this example, we will go over some basic pre-processing of some example text. We use an example text that generated from the GPT-2 in the other example.

```python
text='That data analytics course was a big deal. It was a big deal for the company. It was a big deal for the people in Silicon Valley. It was a big deal for the people who were trying to get in. It is now one of the most popular courses in tech, and it is one of the most talked about courses around. Im not sure how you get to this. But it is a big deal for the tech world, and its not just for me. It is for everyone. What do you think about the impact of the new course on the industry? I think the biggest impact on tech is in terms of the quality of the information that is available. The quality of our information is going to change. Its going to be very interesting to learn about the things that are being done in Silicon Valley today. It is going to be interesting to see what is happening in the next five to ten years.'
text
```

Importing dependencies.

```python
import nltk

from nltk import sent_tokenize
from nltk import word_tokenize
from nltk import FreqDist

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

First, we tokenize the text. This can find us either sentences or words.

```python
sentences = sent_tokenize(text)
sentences
```

```python
words = word_tokenize(text)
words
```

We can find the distribution of words as follows.

```python
dist = FreqDist(words)
dist
```

Let's now remove punctuations:

```python
print(len(words)) # number of terms in the original text document

words_no_punc = []

for w in words:
  if w.isalpha():
    words_no_punc.append(w.lower())

print(words_no_punc)

print(len(words_no_punc))
```

Next, we remove the stopwords. First, let's see what is included in the stopwords.

```python
from nltk.corpus import stopwords
stopwords = stopwords.words("english")
print(stopwords)
```

Now remove them.

```python
clean_words = []

for w in words_no_punc:
  if w not in stopwords:
    clean_words.append(w)

print(clean_words)

print(len(clean_words))
```

Frequency of terms in the cleaned dataset:

```python
dist = FreqDist(clean_words)
dist
```

Let's now create a word cloud.

```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud().generate(text)

plt.figure(figsize = (5,5))
plt.imshow(wordcloud)
```

Then, moving on to stemming.

There are many other types of stemming, you can find some examples in [here](https://towardsai.net/p/nlp/natural-language-processing-nlp-with-python-tutorial-for-beginners-1f54e610a1a0).

We use two different stemming methods.

First, using the [Porter stemmer approach](https://tartarus.org/martin/PorterStemmer/).

```python
from nltk.stem import PorterStemmer

for w in clean_words:
  print(PorterStemmer().stem(w))
```

Now using Snowball stemmer, which uses the same approach as Porter but with some improvement. As well, it is available for several other languages.

```python
from nltk.stem.snowball import SnowballStemmer

for w in clean_words:
  print(SnowballStemmer(language='english').stem(w))
```

We can also get the parts of speech of terms.

```python
nltk.download('averaged_perceptron_tagger_eng')
nltk.pos_tag(clean_words)
```

We can also lemmatize the text. This is similar to stemming, but leaves the word meaningful. For example, whereas the stemm of scarves is scarv, the lemmatization of scarves is scarf:

```python
SnowballStemmer(language='english').stem("scarves")
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

WordNetLemmatizer().lemmatize("scarves")
```

```python
lemmatized_words = [WordNetLemmatizer().lemmatize(word) for word in clean_words]
lemmatized_words
```

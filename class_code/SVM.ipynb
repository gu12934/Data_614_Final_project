# SVM

Here is the code for running SVM in python using the sklearn's SVM function. Some of the code is based on the following references:
https://scikit-learn.org/stable/modules/svm.html

https://www.datacamp.com/tutorial/svm-classification-scikit-learn-python

First, import the required libraries.

```python
import pandas as pd
import numpy as np
from sklearn import svm
```

We then upload our files to Google Colab and then read the files to objects.

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
train = pd.read_csv(io.BytesIO(uploaded['train.csv']))
test = pd.read_csv(io.BytesIO(uploaded['test.csv']))
```

```python
train.shape
```

```python
test.shape
```

Now we are ready to separate the files to training and test inputs and outputs.

```python
trainIn = np.array(train[train.columns[:2308]])
testIn = np.array(test[test.columns[:2308]])
trainOut = np.array(train[train.columns[-1]])
testOut = np.array(test[test.columns[-1]])
```

Now we will create our model. Given that our target variable has multiple classes (4 in this case), we need to use an approach for the classification. The choices are “one-versus-one” classifier (ovo) or a “one-vs-rest” (ovr) classifier.

```python
model = svm.SVC(kernel = 'linear', decision_function_shape='ovo')
model.fit(trainIn, trainOut)
```

We can find the support vectors as:

```python
model.support_vectors_
```

And we can make predictions for the test set as:

```python
model.predict(testIn)
```

We can now calculate the accuracy as:

```python
np.mean(model.predict(testIn) == testOut)
```

We can also derive the probabilistic class assignments if we set the option 'probability' to True:

```python
model2 = svm.SVC(kernel = 'linear', decision_function_shape='ovo', probability = True)
model2.fit(trainIn, trainOut)
```

```python
model2.predict_proba(testIn)
```

There is also an option for giving more importance to some samples if the data is unbalanced.

```python
model3 = svm.SVC(kernel = 'linear', decision_function_shape='ovo', probability = True, class_weight='balanced')
model3.fit(trainIn, trainOut)
```

```python
model3.predict(testIn)
```

```python
np.mean(model3.predict(testIn) == testOut)
```

In our case, this doesn't impact the accuracy.

Generally, SVM is more suitable for classification problems, but it can also be used for regression (numerical output variable). *sklearn* allows for regression modeling as well. You would need to use svm.SVR() (instead of svm.SVC).
